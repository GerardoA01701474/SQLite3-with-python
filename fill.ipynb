```python
from faker import Faker
fake = Faker()
names = [fake.name().split() for i in range(100)]
names = [name for name in names if len(name)==2]
```

```python
import sqlite3 
connection = sqlite3.connect('sample.db')
cursor = connection.cursor()
```

```python
insert_query = 'INSERT INTO people(NAME, surname) VALUES(?,?)' # the python library uses question mark "?" to execute queries that will insert values
for name in names:
    cursor.execute(insert_query, (name[0], name[1]))
connection.commit()
```

```python
select_query = 'SELECT * from people'
for elem in cursor.execute(select_query):
    print(elem)
```
